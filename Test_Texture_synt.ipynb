```python
%load_ext autoreload
%autoreload 2

import Texture_synt as synt
import matplotlib.pyplot as plt
import numpy as np
```

# Example with tomatoes

```python
W = [[9],[17,9],[17,9]]
E = [[2],[4,2],[4,2]]
Res = [4,2,1]
path = "Images/tomatoes.bmp"
Input_im = plt.imread(path).astype(float)/255 
```

```python
Text_synt, Energies, Indices_reschange = synt.multi_level_synth(path,W,E,Res,128)
```

```python
plt.subplots(1,2, figsize = (10,8))
plt.subplot(1,2,1)
plt.imshow(Input_im)
plt.title('Input texture')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(Text_synt.astype(float)/255)
plt.title('Synthetized texture')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(np.arange(len(Energies)),Energies,color = 'tab:blue')
Indices_reschange[2] -= 1
Indices_reschange[-1] -= 1
plt.plot(Indices_reschange,np.array(Energies)[Indices_reschange],'*',markersize = 10, color = 'tab:orange')
plt.xlabel('Iterations')
plt.ylabel('Engergy')
plt.title('Multi-scale synthesis for tomato texture')
plt.show()
```
